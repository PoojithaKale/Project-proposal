```python
import pandas as pd # import and manage datasets
import numpy as np # mathematic library
```

```python
df_investment_tech = pd.read_csv("investment-in-renewable-energy-by-technology.csv")
```

```python
df_investment_tech = df_investment_tech.set_index('Year')
```

```python
df_investment_tech.head()
```

```python
year = range(2004,2017)
total_energy = []
per_solar = []
per_wind = []
per_solar_wind = []
per_marine = []
per_geo = []
per_hydro = []
per_biofuel = []
per_biomass = []
for i in year:
    X = df_investment_tech.loc[i]['Marine Energy (USD)'] + df_investment_tech.loc[i]['Geothermal Energy (USD)'] + df_investment_tech.loc[i]['Small Hydropower (USD)'] + df_investment_tech.loc[i]['Liquid Biofuels (USD)'] + df_investment_tech.loc[i]['Biomass & Waste-to-Energy (USD)'] + df_investment_tech.loc[i]['Wind Energy (USD)'] + df_investment_tech.loc[i]['Solar Energy (USD)']
    Y = df_investment_tech.loc[i]['Solar Energy (USD)']*100/X
    Z = df_investment_tech.loc[i]['Wind Energy (USD)']*100/X
    A = df_investment_tech.loc[i]['Marine Energy (USD)']*100/X
    B = df_investment_tech.loc[i]['Geothermal Energy (USD)']*100/X
    C = df_investment_tech.loc[i]['Small Hydropower (USD)']*100/X
    D = df_investment_tech.loc[i]['Liquid Biofuels (USD)']*100/X
    E = df_investment_tech.loc[i]['Biomass & Waste-to-Energy (USD)']*100/X
    
    total_energy.append(X)
    per_solar.append(Y)
    per_wind.append(Z)
    per_marine.append(A)
    per_geo.append(B)
    per_hydro.append(C)
    per_biofuel.append(D)
    per_biomass.append(E)
    per_solar_wind.append(Y + Z) 
```

```python
df_investment_tech['total_energy'] = total_energy
df_investment_tech['per_solar'] = per_solar
df_investment_tech['per_wind'] = per_wind
df_investment_tech['per_solar_wind'] = per_solar_wind
df_investment_tech['per_marine'] = per_marine
df_investment_tech['per_geo'] = per_geo
df_investment_tech['per_hydro'] = per_hydro
df_investment_tech['per_biofuel'] = per_biofuel
df_investment_tech['per_biomass'] = per_biomass
```

```python
df_investment_tech.head()
```

```python
df_investment_tech.to_excel('Investment_update.xlsx')
```
