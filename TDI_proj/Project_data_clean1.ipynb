```python
df_oil = pd.read_excel(excel_file, 'Oil Consumption - Mtoe') # --- energy from oil consumed 
df_oil_W = df_oil[109:110]
oil_energy = df_oil_W.values.tolist()[0][1:55]
```

```python
df_gas = pd.read_excel(excel_file, 'Gas Consumption - Mtoe') # --- energy from gas consumed 
df_gas_W = df_gas[109:110]
gas_energy = df_gas_W.values.tolist()[0][1:55]
```

```python
df_coal = pd.read_excel(excel_file, 'Coal Consumption - Mtoe') # --- energy from coal consumed 
df_coal_W = df_coal[109:110]
coal_energy = df_coal_W.values.tolist()[0][1:55]
```

```python
df_nuc = pd.read_excel(excel_file, 'Nuclear Consumption - Mtoe') # --- energy from Nuclear consumed 
df_nuc_W = df_nuc[109:110]
nuc_energy = df_nuc_W.values.tolist()[0][1:55]
```

```python
ren_energy = []  # --- energy from renewable sources
for i in range(len(year)):
    x = total_energy[i] - oil_energy[i] - gas_energy[i] - coal_energy[i] - nuc_energy[i]
    ren_energy = ren_energy + [x]
```

```python
oil_gas = []
for i in range(len(year)):
    x = oil_energy[i] + gas_energy[i]
    oil_gas = oil_gas + [x]
```

```python
oil_gas_coal = []
for i in range(len(year)):
    x = oil_energy[i] + gas_energy[i] + coal_energy[i]
    oil_gas_coal = oil_gas_coal + [x]
```

```python
oil_gas_coal_nuc = []
for i in range(len(year)):
    x = oil_energy[i] + gas_energy[i] + coal_energy[i] + nuc_energy[i]
    oil_gas_coal_nuc = oil_gas_coal_nuc + [x]
```

```python
plt.fill_between(year, oil_energy, color='y', alpha=0.5, label = 'Oil' )
plt.fill_between(year, oil_energy, oil_gas,color='g', alpha=0.5, label = 'Gas' )
plt.fill_between(year, oil_gas, oil_gas_coal, color='r', alpha=0.5,label = 'Coal')
plt.fill_between(year, oil_gas_coal, oil_gas_coal_nuc, color='m', alpha=0.5, label = 'Nuclear')
plt.fill_between(year, oil_gas_coal_nuc, total_energy, color='b', alpha=0.5, label = 'Renewable')
plt.xlabel('Years')
plt.xlim((1965, 2018))
plt.ylabel('Million tonnes of oil equivalent (Mtoe)')
plt.legend(loc='upper left')
plt.title('Distribution of energy among the different sources')
plt.show()
```

```python
data_c
```

```python
df_C = pd.DataFrame(data_c, columns = ['Date', 'A2E'+'Afghanistan']) 
```

```python
df_C.head()
```

```python
df_C['Date'] = pd.to_datetime(df_C['Date'])
df_C = df_C.set_index('Date')
```

```python
df_C.head()
```

```python
df_C = df_C.assign(InterpolateLinear =df_C['A2E'+'Afghanistan'].interpolate(method='linear', limit_direction='backward'))
```

```python
df_C = df_C.assign(InterpolateSpline =df_C['A2E'+'Afghanistan'].interpolate(method='spline', order = 1, limit_direction='backward'))
```

```python
df_C = df_C.assign(InterpolateTime = df_C['A2EAfghanistan'].interpolate(method='time', limit_direction='backward'))
```

```python
df_C = df_C.assign(InterpolateAkima =df_C['A2E'+'Afghanistan'].interpolate(method='akima', limit_direction='backward'))
```

```python
# imputing using the rolling average
df_C  = df_C.assign(RollingMean=df_C['A2E'+'Afghanistan'].fillna(df_C['A2E'+'Afghanistan'].rolling(window = 16 ).mean()))
```

```python
# imputing using the rolling median
df_C = df_C.assign(RollingMedian=df_C['A2E'+'Afghanistan'].fillna(df_C['A2E'+'Afghanistan'].rolling(window = 16).median()))
```
