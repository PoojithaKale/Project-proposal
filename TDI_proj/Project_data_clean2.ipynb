```python

df_investment_tech = pd.read_csv("investment-in-renewable-energy-by-technology.csv")
```

```python
df_investment_tech
```

```python
df_investment_tech = df_investment_tech.set_index('Year')
```

```python
year = range(2004,2017)
total_energy = []
per_solar = []
per_wind = []
per_solar_wind = []
for i in year:
    X = df_investment_tech.loc[i]['Marine Energy (USD)'] + df_investment_tech.loc[i]['Geothermal Energy (USD)'] + df_investment_tech.loc[i]['Small Hydropower (USD)'] + df_investment_tech.loc[i]['Liquid Biofuels (USD)'] + df_investment_tech.loc[i]['Biomass & Waste-to-Energy (USD)'] + df_investment_tech.loc[i]['Wind Energy (USD)'] + df_investment_tech.loc[i]['Solar Energy (USD)']
    Y = df_investment_tech.loc[i]['Solar Energy (USD)']*100/X
    Z = df_investment_tech.loc[i]['Wind Energy (USD)']*100/X
    total_energy.append(X)
    per_solar.append(Y)
    per_wind.append(Z)
    per_solar_wind.append(Y + Z) 
```

```python
df_investment_tech['total_energy'] = total_energy
df_investment_tech['per_solar'] = per_solar
df_investment_tech['per_wind'] = per_wind
df_investment_tech['per_solar_wind'] = per_solar_wind
```
