```python
# input is total_investment,investment_per_tech,population,year -> output is growth in RE generation
X1 = pd.DataFrame(list(range(2004,2018)), columns = ['Year'])
X1['T_invest'] = list(India_invest['T_invest_pred'])[:14]
X1['Solar_invest'] = list(India_invest['solar_invest_pred'])[:14]
X1['Population'] = list(India_energy['Population'])[11:]
Y1 = np.log(list(India_RE_gen['Solar Generation TWh'])[14:-1])

X1_train = X1[:10]
X1_test = X1[10:]
Y1_train = Y1[:10]
Y1_test = Y1[10:]
```

```python
X1
```

```python
forcast_X1 = pd.DataFrame(list(range(2017,2031)), columns = ['Year'], index = list(range(24,38)))
```

```python
forcast_X1['T_invest'] = list(India_invest['T_invest_pred'])[13:]
forcast_X1['Solar_invest'] = list(India_invest['solar_invest_pred'])[13:]
forcast_X1['Population'] = list(pop_growth['pop_val_predictions'][57:])
```

```python
from sklearn.linear_model import Ridge

elec_pred1 = Ridge(alpha = 0.11)
elec_pred1.fit(X1_train,Y1_train)
y1_pred = elec_pred1.predict(X1_test)
from sklearn import metrics
print("R^2:", metrics.r2_score(y1_pred, Y1_test))
```

```python
y2_pred = elec_pred1.predict(forcast_X1)
```

```python
import matplotlib.pyplot as plt
plt.plot(X1['Year'], Y1, label='training data')
plt.plot(X1_test['Year'], y1_pred, label='test_predict data')
plt.plot(forcast_X1['Year'], y2_pred, '--', label='model prediction')
plt.legend()
plt.title('Logarithmic growth of Solar')
plt.show()
```
